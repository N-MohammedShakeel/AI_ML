```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split # Sklearn package's randomized data splitting function
```

```python
mydata = pd.read_csv("/content/Housing.csv")
mydata.shape
```

```python
categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea','furnishingstatus']
mydata_encoded = pd.get_dummies(mydata, columns=categorical_cols, drop_first=True ,dtype = int)
```

```python
mydata_encoded.tail()
```

```python
mydata_encoded.dtypes
```

```python
correlation_matrix = mydata_encoded.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
# Based on the heatmap observation: Unwanted or redundant columns
columns_to_drop = ['bedrooms', 'stories']
# Drop the columns using pandas drop
mydata_cleaned = mydata_encoded.drop(columns=columns_to_drop, axis=1)
```

```python
mydata_cleaned.head()
```

```python
mydata_attr = mydata_cleaned.iloc[:, 0:7]
sns.pairplot(mydata_attr, diag_kind='kde')   # to plot density curve instead of histogram on the diag
```

```python
from sklearn.model_selection import train_test_split

X = mydata_cleaned.drop('price', axis=1)
y = mydata_cleaned['price']  # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[idx]))
```

```python
regression_model.score(X_train, y_train)
```

```python

regression_model.score(X_test, y_test)
```
